```python
import numpy as np
import matplotlib.pyplot as plt
```

#### 1. Pthreads
a) mutexes, busy waiting, semaphore

```python
data_mut = np.loadtxt('out_pthread_mutex.txt', delimiter='\t')
data_busy_wait = np.loadtxt('out_busy_wait.txt', delimiter='\t')
data_semaphore = np.loadtxt('out_pthread_semaphore.txt', delimiter='\t')

plt.figure(figsize=(10, 8))
plt.rcParams['font.size'] = 15.0
num_threads = np.arange(len(data_mut)) + 1
plt.plot(num_threads, data_mut[0, 1]/data_mut[:, 1], '.-', label='mutex')
plt.plot(num_threads, data_busy_wait[0, 1]/data_busy_wait[:,1], '.-', label='busy waiting')
plt.plot(num_threads, data_semaphore[0, 1]/data_semaphore[:,1], '.-', label='semaphore')
plt.ylabel('Speedup', fontsize=15)
plt.xlabel('Number of threads')
plt.legend(loc='lower right')
plt.savefig('plot_mutex.png')
plt.show()
```
